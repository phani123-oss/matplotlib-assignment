```python
#Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
#Matplotlib.

'''Matplotlib is exceptionally fast at a variety of operations to make easy to understand 
hard things in a simple way. In addition, it can export visualizations to all popular 
image formats, including PDF, SVG, JPG, PNG, BMP, and GIF. It can create line graphs, 
scatter plots, histograms,bar charts, error charts, pie charts, box plots, and many other
visualization styles.'''
#importing matplotlib module using pyplot submodule
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(5)
y = np.random.rand(len(x))
```

```python
x
```

```python
#Examples of some diff graphs 
#below graphs we have the foloowing same conditions
#1,plotting of corresponding graph
#2.representing labels,tile
#3.initialising figure size
plt.plot(x,y, '--r',alpha=.9)
plt.xlabel("Random values of x")
plt.ylabel("Random values of y")
plt.title("Plot Graph example")
plt.figure(figsize=(4,3))
```

```python
plt.scatter(x,y)
plt.xlabel("Random values of x")
plt.ylabel("Random values of y")
plt.title("Scatter Graph example")
plt.figure(figsize=(4,3))
```

```python
x = (1,1,1,1,1,2,2,2,3,3,3,4,4,4,4,5,5,6,6,6,7,7,7,7,8)
plt.hist(x, color='#36113C')
plt.xlabel("values of x")
plt.ylabel("Frequency of x")
plt.title("Histogram example")
plt.figure(figsize=(4,3))
```

```python
x = ['a','b','c','d','e']
y=[1,2,3,4,5]
plt.bar(x,y)
plt.xlabel("values of x")
plt.ylabel("values of x")
plt.title("Bargraph example")
plt.grid()
plt.figure(figsize=(4,3))
```

```python
x = ['a','b','c','d','e']
y=[1,2,3,4,5]
plt.barh(x,y)
plt.xlabel("values of x")
plt.ylabel("values of x")
plt.title("Bargraph example")
plt.grid()
plt.figure(figsize=(4,3))
```

```python
x = [1,2,3]
y = [4,5,6]
z = [7,8,9]
fig = plt.figure()
#adding subplot 3d using fig variable
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, color='#E333FF',alpha=.9)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.title("3d graph")
plt.show()
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data'''
#given data
import numpy as np
#For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
#For line 2:
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
#For line 3:
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
#For line 4: 
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
print(x1)
print(y1)
print(x2)
print(y2)
print(x3)
print(y3)
print(x4)
print(y4)
```

```python
#The subplot() function takes three arguments that describes the layout of the figure.
#The layout is organized in rows and columns, which are represented by the first and second argument.
#The third argument represents the index of the current plot.
plt.grid()
plt.figure(figsize=(4,3))
plt.subplot(211)
plt.plot(x1,y1)

plt.grid()
plt.figure(figsize=(4,3))
plt.subplot(212)
plt.plot(x2,y2, marker='x')

plt.show()
```

```python
plt.plot(x3,y3, marker='*' ,color='g')
plt.grid()
plt.figure(figsize=(4,3))
plt.subplot(121)


plt.plot(x4,y4, marker='*', color='r')
plt.grid()
plt.figure(figsize=(4,3))
plt.subplot(212)

plt.show()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#given data
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

#Bar graph can be drawn between strings and numeric values
#horizontal bargraph is the same process of vertical bar graph by using 
#barh vertical graph changes to horizontal position
```

```python
plt.bar(company,profit, color='#E333FF' )
plt.xlabel('company names')
plt.ylabel('profits')
plt.title("About company name and profit using Bar Graph")
plt.figure(figsize=(2,3))
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel('company names')
plt.ylabel('profits')
plt.title("About company name and profit using Horizontal Bar Graph")
plt.figure(figsize=(2,3))
plt.show()
```
